```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
# Load dataset
ins_data = pd.read_csv('insurance.csv')
ins_data
```

```python
# Basic overview
ins_data.info()
```

```python
# Checking for missing values
ins_data.isna().sum()
```

```python
# Checking for duplicated rows
ins_data.duplicated().sum()
```

```python
# Drop duplicates
ins_data.drop_duplicates(inplace=True)
```

```python
cat_cols = ins_data.select_dtypes('object').columns
num_cols = ins_data.select_dtypes(exclude='object').columns.difference(['charges'])
target_col = ['charges']
```

```python
plt.figure(figsize=(4,3))
corr = ins_data[num_cols].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation", fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(9,7))
for index, i in enumerate(num_cols):
    plt.subplot(2,2, index+1)
    sns.histplot(x=i, data=ins_data, kde=True)
    plt.title(f"Distribution of {i}", fontweight='bold')
plt.tight_layout()
```

```python
plt.figure(figsize=(9,7))
for index, i in enumerate(num_cols):
    plt.subplot(2,2, index+1)
    sns.histplot(x=i, data=ins_data, kde=True, hue='sex')
    plt.title(f"Distribution of {i}", fontweight='bold')
plt.tight_layout()
```

```python
plt.figure(figsize=(9,7))
for index, i in enumerate(num_cols):
    plt.subplot(2,2, index+1)
    sns.histplot(x=i, data=ins_data, kde=True, hue='smoker')
    plt.title(f"Distribution of {i}", fontweight='bold')
plt.tight_layout()
```

```python
plt.figure(figsize=(8,6))
for index, i in enumerate(cat_cols):
    plt.subplot(2,2,index+1)
    sns.countplot(x=i, data=ins_data)
    plt.title(f"Distribution of {i}", fontweight='bold')
plt.tight_layout()
```

```python
plt.figure(figsize=(8,6))
for index, i in enumerate(cat_cols):
    plt.subplot(2,2,index+1)
    sns.boxplot(x=i,y='charges', data=ins_data)
    plt.title(f"Distribution of {i}", fontweight='bold')
plt.tight_layout()
```
